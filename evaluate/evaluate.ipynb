```python
gold_psalms = "https://docs.google.com/spreadsheets/d/1YRBLiZqsTPZEBiyuZ01-gKxcPyl9VT0IoNE_07X0Hsg/edit?gid=0"
gold_gospels = "https://docs.google.com/spreadsheets/d/1xppiRNK4cCHHW3GqDCL8tliAeZDPg87PTfME80rnvE4/edit?gid=0"
gold_clozianus1 = "https://docs.google.com/spreadsheets/d/1hBilgcFMrdbnLkqM2x21ke42WZaAbyFJ/edit?gid=945687746"
gold_clozianus2 = "https://docs.google.com/spreadsheets/d/1hBilgcFMrdbnLkqM2x21ke42WZaAbyFJ/edit?gid=1370115719"
surl = gold_psalms
surl = gold_clozianus2
```

```python
ex = Extractor.build(surl)
ex.df.head()
```

```python
if os.path.isfile(ex.fname):
    os.remove(ex.fname)
```

```python
for test, expected, src in tqdm(ex):
    df = await evaluate(test, expected, src, ex.fname)
    if df.empty:
        print(f"Empty: {test}, {expected}, {ex.fname}")
        continue
    if os.path.isfile(ex.fname):
        res = [pd.read_excel(ex.fname)] + [df]
    else:
        res = [df]
    df2 = pd.concat(res, axis=0)
    df2.to_excel(ex.fname, index=False)
```
