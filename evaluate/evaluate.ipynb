# Show examples from spreadsheet

```python
ex = Extractor.build(surl)
ex.df.head()
```

```python
print(f"Deleting {ex.fname} if previously saved.")
if os.path.isfile(ex.fname):
    os.remove(ex.fname)
```

# Query algorithms with examples and save evaluation

```python
for test, expected, src in tqdm(ex):
    df = await evaluate(test, expected, src, ex.fname)
    if df.empty:
        print(f"Empty: {test}, {expected}, {ex.fname}")
        continue

    # examples are added to file as they are calculated, so before end the file exists but is incomplete.
    if os.path.isfile(ex.fname):
        res = [pd.read_excel(ex.fname)] + [df]
    else:
        res = [df]
    df2 = pd.concat(res, axis=0)
    df2.to_excel(ex.fname, index=False)
```
